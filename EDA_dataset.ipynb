```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
df = pd.read_csv("sample_fr_en_clean.csv")
```

```python
df
```

```python
# Aper√ßu des 5 premi√®res lignes
print("üìå Aper√ßu du dataset :")
print(df.head())

# Infos g√©n√©rales
print("\nüß† Infos g√©n√©rales :")
print(df.info())

# Dimensions du dataset
print(f"\nüìê Taille du dataset : {df.shape[0]} lignes, {df.shape[1]} colonnes")

# V√©rification des valeurs manquantes
print("\nüîç Valeurs manquantes par colonne :")
print(df.isnull().sum())

# Statistiques sur la longueur des phrases
df['fr_len'] = df['fr'].str.len()
df['en_len'] = df['en'].str.len()

print("\nüìä Statistiques des longueurs de phrases :")
print(df[['fr_len', 'en_len']].describe())
```

```python
plt.figure(figsize=(12, 5))

# Longueur phrases fran√ßaises
sns.histplot(df['fr_len'], bins=50, color='skyblue', label='Fran√ßais', kde=True)

# Longueur phrases anglaises
sns.histplot(df['en_len'], bins=50, color='salmon', label='Anglais', kde=True)

plt.title("Distribution des longueurs de phrases")
plt.xlabel("Nombre de caract√®res")
plt.ylabel("Fr√©quence")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Nuage de mots - fran√ßais
text_fr = ' '.join(df['fr'].tolist())
wordcloud_fr = WordCloud(width=800, height=400, background_color='white').generate(text_fr)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_fr, interpolation='bilinear')
plt.axis('off')
plt.title("Nuage de mots - Fran√ßais")
plt.show()
```
